# Consultando um dataset
- Podemos fazer consultas em um Dataframe(Parecido com SQL).
- Existem métodos de consulta usando operadores lógicos (>,<,==)
- Podemos consultar usando instruções de agrupamento.
- Entre outros

```python
import pandas as pd
```

```python
arquivo = 'Arquivos/kc_house_data.csv'
```

```python
dataset = pd.read_csv(arquivo, sep=',')
```

```python
dataset.head()
```

## DataFrame.value_counts(subset=None, normalize=False, sort=True, ascending=False)
### Return a Series containing counts of unique rows in the DataFrame. 

Source: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.value_counts.html?highlight=value_counts

```python
# Calcula a quantidade de valores únicos da coluna
pd.value_counts(dataset['bedrooms'])
```

## property DataFrame.loc
### Access a group of rows and columns by label(s) or a boolean array.

Source: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.loc.html?highlight=loc

```python
# Método loc() é utilizado para verificar informações de um dataset
# O método recebe uma lista por parâmetro e retorna o resultado da consulta
# Consultar imóveis com 3 quartos: Total de 9822 resultados.
dataset.loc[dataset['bedrooms']==3]
```

```python
# No método loc() podemos utilizar em conjunto com o operador &
# Consultar imóveis com 3 quartos e banheiros > 2: Total de 356 registros.
dataset.loc[(dataset['bedrooms']==2) & (dataset['bathrooms'] > 2)]
```

## DataFrame.sort_values(by, axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last', ignore_index=False, key=None)
### Sort by the values along either axis.

```python
# Método sort_values() ordena o dataset
# Retorna a query ordenanda e não o dataset em memória
dataset.sort_values(by='price', ascending=False)
```

```python
# Count() retorna o número de linhas a partir de uma condição.
dataset[dataset['bedrooms']==4].count()
# Entendendo a exeução da instrução
# Primeiramente irá retornar todo o dataset (primeiro dataset escrito)
# Após isso vai aplicar o filtro: dataset['bedrooms']==4, ou seja, retornar do dataset completo apenas os que tem 4 quartos
# Após resultado dessa query, realiza o count() das linhas
```

# Alterando Dataset

```python
# Adicionando uma nova coluna do Dataframe
dataset['size'] = (dataset['bedrooms']*20)
# Ao passar esse valor dataset['size'], pandas entende que não existe e cria essa coluna
```

```python
# Visualizando o conteudo da coluna criada
dataset['size'].head()
```

```python
# Criando uma função para processamento de Dados
def categoriza(s):
    if s>=80:
        return 'Big'
    elif s>=60:
        return 'Medium'
    elif s>=40:
        return 'Small'
```

```python
# Apply() para cada linha de uma determinada coluna, a função é chamada passando como parametro o valor da linha
dataset['cat_size'] = dataset['size'].apply(categoriza)
```

```python
dataset['cat_size'].head()
```

```python
# Tranformando em tipo category para manter os bons padrões de economia de recurso.
dataset.cat_size = dataset.cat_size.astype('category')
```

```python
dataset['cat_size'].head()
```

```python
# Ver a distribuição da coluna com o método value_counts
pd.value_counts(dataset['cat_size'])
```

# DataFrame.drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')
Drop specified labels from rows or columns.

Remove rows or columns by specifying label names and corresponding axis, or by specifying directly index or column names. When using a multi-index, labels on different levels can be removed by specifying the level.

source: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html?highlight=drop#pandas.DataFrame.drop

```python
# Drop() é utilizado para excluir dados do dataframe
# Paramatrização axis=1 define que queremos excluir uma coluna e não uma linha ({0 or ‘index’, 1 or ‘columns’}, default 0)
# Parametrização inplace define que a alteração irá modificar o objeto em memória,
# sem precisar ficar reatribuindo o valores do dataset
dataset.drop(['cat_size'], axis=1, inplace=True)
```

```python
dataset.drop(['size'], axis=1, inplace=True)
```

```python
dataset.head()
```

### Apagando linhas baseado em condições lógicas

```python
#Retorna a lista de index com os valores dos bedrooms = a 0
dataset[dataset.bedrooms==0].index
```

```python
# Dropar linhas com bedrooms = 0
# Esse lista de index é parametro para o drop
dataset.drop(dataset[dataset.bedrooms==0].index, inplace=True)
```

```python
# Dropar linhas maiores que 30
dataset.drop(dataset[dataset.bedrooms>30].index, inplace=True)
```

```python
# retorna a quantidade de valores únicos, podemos ver que não temos mais os valores de quartos maiores que 30.
pd.value_counts(dataset['bedrooms'])
```

# PERCORRENDO LINHAS DE UM DATAFRAME PANDAS
- Método iterrows() permite percorer por todas as linhas de um dataframe
- Método retorna um objeto **iterator** que contém o indice de cada linha e um dado do tipo série (indice, linha)

```python
type(dataset.iterrows()) #Retorna um objeto iterator
```

```python
# Imprime a primeira linha do objeto iterator
next(dataset.iterrows())
```

```python
# Caso queiramos percorrer imprimindo o indice e cada linha
for indice, linha in dataset.head(10).iterrows():
    print(indice,linha)
```

```python
# Percorrendo o dataset acessando o nome das colunas
for indice,linha in dataset.head(10).iterrows():
    print(indice,linha['id'], linha['price'], linha['bedrooms'])    
```

### Atualizando o Dataframe ao percorrer linha a linha 

```python
dataset.price.head()
```

## property DataFrame.at
Access a single value for a row/column label pair.

Similar to loc, in that both provide label-based lookups. Use at if you only need to get or set a single value in a DataFrame or Series.

```python
# Percorrendo e atualizando linhas de um dataframe.
# Atualizar o valor da coluna PRICE multiplicado por 2
# Metodo at() nos ajudará nisso
# Não conseguimos mudar os valores e aplicar no DF diretamente do iterrows() pq ele nos retorna uma tupla, e tupla são imutáveis
for indice, linha in dataset.iterrows():
    dataset.at[indice,'price'] = linha['price'] * 2
#Entendendo o código:
#A partir do dataset: dataset (já muda no dataset)
#Vai pegar o valor do PRICE da linha que está sendo iterada: .at[indice,'price']
#E atualizar o valor a partir do valor retornado do atributo daquela linha multiplicado por 2: = linha['price'] * 2
```

```python
# Valores atualizados
dataset.price.head()
```

### Percorrendo Dataframe com o método itertuples()
- Retorna as linhas e indice em formato de tuplas
- Costuma ser mais rápido que o iterrows()

```python
for linha in dataset.head().itertuples():
    print(linha)
```

```python
# Imprime linhas chamando as colunas
for linha in dataset.head().itertuples():
    print(linha.Index, linha.id, linha.bedrooms, linha.price)
```

```python
# E Se eu quiser atualizar ?
for linha in dataset.head(5).itertuples():
    dataset.at[linha.Index, 'price'] = linha.price * 3
```

```python
for linha in dataset.head(5).itertuples():
    print(linha.Index, linha.id, linha.bedrooms, linha.price)
```

# Missing Values 
- São valores faltantes em colunas, esses podem ser oriundos de falhas em cargas de dados, falhas em crawlers ou até mesmo corrupção de dados
- Podem ser um problema em várias soluções, como por exemplo, algoritmos de machine learning que não trabalham bem com dados faltantes.
- Podem atrapalhar resultados de análises.
- Como identificá-los ?

```python
arquivo = 'Arquivos/kc_house_data.csv'
dataset = pd.read_csv(arquivo, sep=',', header=0)
```

```python
dataset.head()
```

```python
# Consultando linhas com valores faltantes
dataset.isnull()
```

```python
# Consultando a quantidade de linhas com valores faltantes
dataset.isnull().sum()
```

# DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
Remove missing values.

0, or ‘index’ : Drop rows which contain missing values.

1, or ‘columns’ : Drop columns which contain missing value.

```python
# dropna() - este comando remove todas as linhas onda tenha pelo menos um registro faltante me algum atributo.
dataset.dropna(inplace=True) #lembrando que inplace true já atualiza o do dataset na memória.
```

```python
#Confirmando que foi removido
dataset.isnull().sum()
```

```python
# Refazendo o dataset para demais testes
dataset = pd.read_csv(arquivo, sep=',', header=0)
```

# how{‘any’, ‘all’}, default ‘any’
Determine if row or column is removed from DataFrame, when we have at least one NA or all NA.

**‘any’** : If any NA values are present, drop that row or column.

**‘all’** : If all values are NA, drop that row or column.

```python
# Remover somente linhas qu estejam com valores faltantes em TODAS as colunas
dataset.dropna(how='all', inplace=True)
```

```python
#Confirmando que foi removido, devido passarmos o atributo ALL e nosso dataset não conter registros com valores faltante
# em todas as colunas, logo nada foi removido.
dataset.isnull().sum()
```

```python
# Outra opção de trablhar com valores nulos é atribuir um valor para esses valores faltantes
#Preenche com a média dos valores da coluns floors os values null
dataset['floors'].fillna(dataset['floors'].mean(), inplace=True)
```

```python
# preencher com valor 1 os valores null da coluna bedroom
dataset['bedrooms'].fillna(1, inplace=True)
```

```python
dataset.isnull().sum()
```
