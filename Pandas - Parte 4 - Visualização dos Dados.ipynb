# Visualização dos Dados

```python
import pandas as pd
```

```python
arquivo = 'Arquivos/kc_house_data.csv'
dataset = pd.read_csv(arquivo, sep=',')
```

```python
dataset.head()
```

```python
# Plotar um histograma com a coluna Price com 30 bins na cor vermelha
# 30 bins significa que vamos agrupar os dados em 30 (na prática dividido em 30 barras), se diminuirmos mais separadão fica.
dataset['price'].hist(bins=30, color='red')
```

```python
# Mesma chamada só que com mais bins
dataset['price'].hist(bins=50, color='blue')
```

```python
# Mesma chamada só que com menos bins
dataset['price'].hist(bins=10, color='black')
```

```python
# Plotar 2 colunas lado a lado na cor verde
dataset[['bedrooms','bathrooms']].hist(bins=30, alpha=0.5, color='Green') # alpha = Opacidade
```

### Estatística Descritiva 

```python
# Valor médio, de uma coluna (parametro do dataset)
dataset['bedrooms'].mean()
```

```python
# Valor Máximo
dataset['bedrooms'].max()
```

```python
# Valor Mínimo
dataset['bedrooms'].min()
```

```python
# Desvio padrão
dataset['bedrooms'].std()
```

```python
# Retorna a simetria da coluna
dataset['bedrooms'].skew()
```

```python
# Valor zero indica uma distribuição simétrica
# Valor maior que zero ou menor indica uma distribuição assimétrica.
# Valores acima do zero podemos dizer que existe uma assimetria positiva.
# valores abaixo de zero uma assimetria negativa
# Isso quer dizer que valores muito acima de zero indicam que existe mmais valores acima da média
# E valores baixo de zero significa que contém mais valores abaixo da média.
dataset.skew()
```

# Correlação
- Apresentam o nível de dependência de uma varíavel está relacionada com a outra
- Positiva: Quando uma aumenta a outra tb aumenta
- Negativa: Quando uma aumenta a outra diminui e vice-versa.

```python
# Imprime a correlação de todas as colunas do dataframe (PERSON)
dataset.corr()
```

```python
# Imprime a correlação de spearman(esta leva em considerção a correção positiva e negativa)
dataset.corr('spearman')
```

```python
# Imprime a correlação de algumas colunas apenas.
dataset[['price','bedrooms','bathrooms','floors']].corr()
```

```python
# Plotando a correlação de algumas colunas
%matplotlib notebook
dataset[['price','bedrooms','bathrooms','floors']].corr().plot()
```

# Tabelas PIVOT
- São úteis para agrupamentos de dados

```python
# Conta a qualidade de imóveis agrupados pelas colunas waterfront e floors
# parametro INDEX: informa as colunas que serão usadas para agregação
# parametro AGGFUNC: é usado para definir a função de agregação (count, média, etc.)
# parametro MAGINS: Se TRUE calcula a quantidade total no final da tabela

# Muito interessante a performance dessa operação
dataset.pivot_table('id', index=['waterfront','floors'], aggfunc='count', margins=True)
```

# Crosstab ou Tabulação Cruzada
- Cruzar valores das variáveis

```python
# Distribuição de imóveis por número de quartos com relação a condição
# Podemos ver que imoveis com 3 quartos estão mais na condição 3
pd.crosstab(dataset['bedrooms'],dataset['condition'])
```

```python
# Como plotar um gráfico de uma tabulação cruzada
# Grid = false, não vai ter linha no fundo do gráfico
table = pd.crosstab(dataset['bedrooms'],dataset['condition'])
table.plot(kind='bar', width=1.0, color=['red','yellow','orange','blue','green'], title='Condition by Bedrooms',
          grid=False)
```
